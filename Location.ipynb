```python
import pandas as pd
```

```python
restaurant_df = pd.read_csv('data/Restaurant.csv')
hotel_df = pd.read_csv('data/Hotel.csv')
touristSite_df = pd.read_csv('data/Tourist_site.csv')
```

```python
# standardize location variables
# Metro Manila is represented as Manila in the hotel_df
# replace all Manila entries by Metro Manila
hotel_df['Province'].replace('Manila', 'Metro Manila', inplace=True)
```

```python
#  merge relevant collected data
# to concat 
restaurant_features = restaurant_df.copy()[['Restaurant Name', 'Block', 'City', 'Province']]
restaurant_features.insert(1, 'Classification', 'Restaurant')

hotel_features = hotel_df.copy()[['Hotel', 'Block', 'City', 'Province']]
hotel_features.insert(1, 'Classification', 'Hotel')

touristSite_features = touristSite_df.copy()[['name', 'block', 'city', 'province']]
touristSite_features.insert(1, 'Classification', 'Tourist Site')

restaurant_features.rename(columns = {'Restaurant Name': 'Name'}, inplace=True)
hotel_features.rename(columns = {'Hotel': 'Name'}, inplace=True)
touristSite_features.rename(columns = {'name':'Name', 'block':'Block', 'city':'City', 'province':'Province'}, inplace=True)
```

```python
location_df = pd.concat([restaurant_features, hotel_features, touristSite_features], ignore_index=True)
```

```python
# get unique Provinces to use as basis for grouping
unique_provinces = list(location_df.Province.unique())
unique_provinces
```

```python
# handle the missing and wrong values
location_df.at[59, 'City'] = 'El Nido'
location_df.at[59, 'Province'] = 'Palawan' 

location_df.at[61, 'City'] = 'Malay'
location_df.at[61, 'Province'] = 'Aklan' 

location_df.at[50, 'City'] = 'Vigan'
location_df.at[50, 'Province'] = 'Ilocos Sur' 
```

```python
new_unique_provinces = list(location_df.Province.unique())
new_unique_provinces
```

```python
# for each province
# get the relevant rows save the output as csv

for province in new_unique_provinces:
    province_df = location_df[location_df.Province == province]
    province_df.to_csv('data/location_data/{0}_Location.csv'.format(province))
```
